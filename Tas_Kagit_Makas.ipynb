```python
import gradio as gr
import random
```

# Oyun iÃ§in gerekli sayaÃ§lar

```python
# Global skor deÄŸiÅŸkenleri
user_score = 0
computer_score = 0
```

# GeniÅŸletilmiÅŸ oyun seÃ§enekleri ve kurallar


```python
# GeniÅŸletilmiÅŸ oyun seÃ§enekleri ve kurallar
choices = ['TaÅŸ', 'KaÄŸÄ±t', 'Makas', 'Spock', 'Kertenkele']
winning_combinations = {
    'TaÅŸ': ['Makas', 'Kertenkele'],
    'KaÄŸÄ±t': ['TaÅŸ', 'Spock'],
    'Makas': ['KaÄŸÄ±t', 'Kertenkele'],
    'Spock': ['TaÅŸ', 'Makas'],
    'Kertenkele': ['KaÄŸÄ±t', 'Spock']
}
```

# Oyun devam etmeli mi?

```python
def devam_etme_istegi():
    # KullanÄ±cÄ±ya soralÄ±m
    user_response = gr.Radio(["Evet", "HayÄ±r"], label="BaÅŸka bir oyun oynamak ister misiniz?")

    # BilgisayarÄ±n rastgele cevabÄ±
    computer_response = random.choice(["Evet", "HayÄ±r"])

    return user_response, computer_response
```

# TaÅŸ, KaÄŸÄ±t, Makas oyun fonksiyonu





```python
# TaÅŸ, KaÄŸÄ±t, Makas oyun fonksiyonu
def tas_kagit_makas_ozgun(player_choice, difficulty, theme, play_again_user):
    global user_score, computer_score

    # BilgisayarÄ±n rastgele seÃ§imi
    computer_choice = random.choice(choices)

    # Zorluk seviyesine gÃ¶re bilgisayar seÃ§imi
    if difficulty == "Zor":
        if random.random() < 0.7:  # Zor seviyede bilgisayarÄ±n kazanma olasÄ±lÄ±ÄŸÄ±nÄ± artÄ±rma
            for choice in winning_combinations[computer_choice]:
                if choice == player_choice:
                    computer_choice = random.choice(winning_combinations[player_choice])
                    break

    # SonuÃ§larÄ± belirleme
    if player_choice == computer_choice:
        result = "Berabere!"
    elif computer_choice in winning_combinations[player_choice]:
        result = "Bu turu kazandÄ±nÄ±z!"
        user_score += 1
    else:
        result = "Bu turu kaybettiniz!"
        computer_score += 1

    # Skor durumu
    score_display = f"Skor: Oyuncu {user_score} - {computer_score} Bilgisayar"

    # Ã–dÃ¼l sistemi
    reward_message = ""
    if user_score >= 5:
        reward_message = "Tebrikler! 5 galibiyet kazandÄ±nÄ±z! ðŸŽ‰"

    # BilgisayarÄ±n oyuna devam etme isteÄŸi
    play_again_computer = random.choice(["Evet", "HayÄ±r"])

    # KullanÄ±cÄ± ve bilgisayar oyuna devam etmek istiyor mu?
    if play_again_user == "Evet" and play_again_computer == "Evet":
        continue_message = "Hem siz hem de bilgisayar devam etmek istiyorsunuz. Oyun devam ediyor!"
    elif play_again_user == "HayÄ±r" or play_again_computer == "HayÄ±r":
        continue_message = "Oyun bitti. TeÅŸekkÃ¼r ederiz!"
        user_score = 0
        computer_score = 0
    else:
        continue_message = "Oyun devam ediyor!"

    return result, computer_choice, score_display, reward_message, continue_message
```

# Gradio arayÃ¼zÃ¼

```python
# Gradio arayÃ¼zÃ¼ bileÅŸenleri
title = "GeliÅŸmiÅŸ TaÅŸ, KaÄŸÄ±t, Makas Oyunu"
description = "Kurallar: TaÅŸ Makas'Ä± yener, Makas KaÄŸÄ±t'Ä± yener, KaÄŸÄ±t TaÅŸ'Ä± yener. Spock ve Kertenkele seÃ§enekleriyle daha fazla eÄŸlence!"

# KullanÄ±cÄ± girdisi iÃ§in Dropdown
player_choice = gr.Dropdown(choices, label="SeÃ§iminizi YapÄ±n")
difficulty = gr.Radio(["Kolay", "Orta", "Zor"], label="Zorluk Seviyesi")
theme = gr.Radio(["AÃ§Ä±k", "Koyu"], label="Tema SeÃ§imi")
play_again_user = gr.Radio(["Evet", "HayÄ±r"], label="BaÅŸka bir oyun oynamak ister misiniz?")

# Ã‡Ä±ktÄ±lar iÃ§in Textbox
result_text = gr.Textbox(label="Tur Sonucu")
computer_choice_text = gr.Textbox(label="BilgisayarÄ±n SeÃ§imi")
score_text = gr.Textbox(label="Skor Durumu")
reward_text = gr.Textbox(label="Ã–dÃ¼l")
continue_text = gr.Textbox(label="Oyun Durumu")
```

```python
# Gradio arayÃ¼zÃ¼
demo = gr.Interface(
    fn=tas_kagit_makas_ozgun,
    inputs=[player_choice, difficulty, theme, play_again_user],
    outputs=[result_text, computer_choice_text, score_text, reward_text, continue_text],
    title=title,
    description=description,
    theme="default",
    live=True
)

# UygulamayÄ± baÅŸlat
demo.launch(share=True)
```
